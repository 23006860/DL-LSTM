````python
import pandas as pd
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from torch.nn.utils.rnn import pad_sequence
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
# Load and prepare data
data = pd.read_csv("ner_dataset.csv", encoding="latin1").ffill()
words = list(data["Word"].unique())
tags = list(data["Tag"].unique())

if "ENDPAD" not in words:
    words.append("ENDPAD")

word2idx = {w: i + 1 for i, w in enumerate(words)}
tag2idx = {t: i for i, t in enumerate(tags)}
idx2tag = {i: t for t, i in tag2idx.items()}
data.head(50)
print("Unique words in corpus:", data['Word'].nunique())
print("Unique tags in corpus:", data['Tag'].nunique())
print("Unique tags are:", tags)
# Group words by sentences
class SentenceGetter:
    def __init__(self, data):
        self.grouped = data.groupby("Sentence #", group_keys=False).apply(
            lambda s: [(w, t) for w, t in zip(s["Word"], s["Tag"])]
        )
        self.sentences = list(self.grouped)

getter = SentenceGetter(data)
sentences = getter.sentences
sentences[35]
# Encode sentences
X = [[word2idx[w] for w, t in s] for s in sentences]
y = [[tag2idx[t] for w, t in s] for s in sentences]
word2idx
plt.hist([len(s) for s in sentences], bins=50)
plt.show()
# Pad sequences
max_len = 50
X_pad = pad_sequence([torch.tensor(seq) for seq in X], batch_first=True, padding_value=word2idx["ENDPAD"])
y_pad = pad_sequence([torch.tensor(seq) for seq in y], batch_first=True, padding_value=tag2idx["O"])
X_pad = X_pad[:, :max_len]
y_pad = y_pad[:, :max_len]
X_pad[0]
y_pad[0]
# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_pad, y_pad, test_size=0.2, random_state=1)

# Dataset class
class NERDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return {
            "input_ids": self.X[idx],
            "labels": self.y[idx]
        }

train_loader = DataLoader(NERDataset(X_train, y_train), batch_size=32, shuffle=True)
test_loader = DataLoader(NERDataset(X_test, y_test), batch_size=32)
````

````python
# Model definition
class BiLSTMTagger(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim):
        """
        Initializes the BiLSTMTagger model.

        Args:
            vocab_size (int): The size of the vocabulary (number of unique words + padding).
            embedding_dim (int): The dimension of the word embeddings.
            hidden_dim (int): The number of features in the hidden state of the LSTM.
            output_dim (int): The number of output tags.
        """
        super(BiLSTMTagger, self).__init__()

        # Embedding layer: Converts word indices to dense vectors
        self.embedding = nn.Embedding(vocab_size, embedding_dim)

        # BiLSTM layer: Processes the embedded sequence
        # batch_first=True means input and output tensors are (batch, seq, feature)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, bidirectional=True, batch_first=True)

        # Linear layer: Maps LSTM output to tag scores
        # Bi-directional LSTM outputs concatenate hidden states from both directions,
        # so the input feature size to the linear layer is 2 * hidden_dim
        self.fc = nn.Linear(hidden_dim * 2, output_dim)

    def forward(self, input_ids):
        """
        Defines the forward pass of the model.

        Args:
            input_ids (torch.Tensor): Input tensor containing word indices.

        Returns:
            torch.Tensor: Output tensor containing the predicted tag scores for each word.
        """
        # Get embeddings for the input word indices
        embedded = self.embedding(input_ids)

        # Pass the embedded sequence through the BiLSTM
        # lstm_out contains the hidden states for each time step
        # _ is the final hidden and cell states (not needed for this task)
        lstm_out, _ = self.lstm(embedded)

        # Apply the linear layer to the BiLSTM output
        # Reshape the output to apply the linear layer to each word's representation
        # lstm_out has shape (batch_size, seq_len, hidden_dim * 2)
        # We want to apply the linear layer to each (hidden_dim * 2) vector
        output = self.fc(lstm_out)

        # The output has shape (batch_size, seq_len, output_dim)
        # Each element output[i, j, k] represents the score for the k-th tag
        # for the j-th word in the i-th sentence.
        return output
````

````python
model = BiLSTMTagger(len(words) + 1, embedding_dim=128, hidden_dim=256, output_dim=len(tags)).to(device)
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
````

````python
def train_model(model, train_loader, test_loader, loss_fn, optimizer, epochs=3):
    train_losses = [] # Initialize list to store training loss per epoch
    val_losses = []   # Initialize list to store validation loss per epoch

    for epoch in range(epochs):
        # **Training Phase**
        model.train()  # Set the model to training mode. This enables layers like Dropout if used.
        running_loss = 0.0 # Accumulate loss for the current epoch
        for batch in train_loader: # Iterate over batches provided by the DataLoader
            input_ids = batch["input_ids"].to(device) # Get input sequences and move to the specified device (CPU/GPU)
            labels = batch["labels"].to(device)       # Get true labels and move to the device

            optimizer.zero_grad() # Zero the gradients before starting to do backpropagation. This is important as gradients accumulate by default.

            # Forward pass
            outputs = model(input_ids) # Pass the input data through the model to get predictions

            # Calculate the loss
            # We need to reshape the outputs and labels for the loss function.
            # outputs shape: (batch_size, seq_len, output_dim) -> view(-1, output_dim) reshapes to (batch_size * seq_len, output_dim)
            # labels shape: (batch_size, seq_len) -> view(-1) reshapes to (batch_size * seq_len)
            # This flattens all predictions and labels across the batch and sequence length for the loss calculation.
            loss = loss_fn(outputs.view(-1, outputs.shape[-1]), labels.view(-1))

            # Backward pass and optimize
            loss.backward() # Compute gradient of the loss with respect to model parameters
            optimizer.step() # Update the model parameters using the calculated gradients and the optimizer

            # Accumulate loss for the epoch
            # We multiply by input_ids.size(0) to get the loss for the entire batch, not just the average per element.
            running_loss += loss.item() * input_ids.size(0)

        # Calculate average training loss for the epoch
        epoch_train_loss = running_loss / len(train_loader.dataset)
        train_losses.append(epoch_train_loss)
        print(f"Epoch {epoch+1}/{epochs}, Training Loss: {epoch_train_loss:.4f}")

        # **Evaluation Phase**
        model.eval()  # Set the model to evaluation mode. This disables layers like Dropout and uses running averages for BatchNorm if used.
        running_val_loss = 0.0 # Accumulate validation loss for the current epoch
        with torch.no_grad():  # Disable gradient calculation. This saves memory and computation during evaluation as we don't need to backpropagate.
            for batch in test_loader: # Iterate over batches from the test DataLoader
                input_ids = batch["input_ids"].to(device)
                labels = batch["labels"].to(device)

                outputs = model(input_ids) # Get predictions from the model

                # Calculate validation loss, similar to training loss
                loss = loss_fn(outputs.view(-1, outputs.shape[-1]), labels.view(-1))

                running_val_loss += loss.item() * input_ids.size(0)

        # Calculate average validation loss for the epoch
        epoch_val_loss = running_val_loss / len(test_loader.dataset)
        val_losses.append(epoch_val_loss)
        print(f"Epoch {epoch+1}/{epochs}, Validation Loss: {epoch_val_loss:.4f}")

    return train_losses, val_losses # Return the lists of training and validation losses
````

````python
def evaluate_model(model, test_loader, X_test, y_test):
    model.eval()
    true_tags, pred_tags = [], []
    with torch.no_grad():
        for batch in test_loader:
            input_ids = batch["input_ids"].to(device)
            labels = batch["labels"].to(device)
            outputs = model(input_ids)
            preds = torch.argmax(outputs, dim=-1)
            for i in range(len(labels)):
                for j in range(len(labels[i])):
                    if labels[i][j] != tag2idx["O"]:
                        true_tags.append(idx2tag[labels[i][j].item()])
                        pred_tags.append(idx2tag[preds[i][j].item()])
````

````python
# Run training and evaluation
train_losses, val_losses = train_model(model, train_loader, test_loader, loss_fn, optimizer, epochs=3)
evaluate_model(model, test_loader, X_test, y_test)
````
